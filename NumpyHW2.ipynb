### HW. Блок Numpy. Задача 4

- Создайте переменную **array_four** и присвойте ей значение сгенерированной матрицы случайными значениями при помощи numpy **размером 11 строк на 8 столбцов** (должен быть тип переменной numpy.ndarray)
- Получите сингулярное разложение матрицы **array_four** при помощи numpy, запишите в соответствующие переменные **U, S, Vt**, где:
    - **U** - ортогональная матрица
    - **S** - массив, числа которого используются в диагональной матрице $\Sigma$
    - **Vt** - ортогональная матрица в уже транспонированном виде (ничего транспонировать самим не нужно)
- Создайте переменную **result** и присвойте ей значение переменожения **U**, **S** и **Vt**, где от матрицы **u** необходимо взять по 8 строк и 8 столбцов (подсказка U[:8, :8]@ ....)

```python
# Ваш код

array_four =  np.random.randint(low=-100, high=100, size=(11, 8))
print("array_four")
print(array_four)
U, S, Vt = np.linalg.svd(array_four)
print("\nОртогональная матрица U:")
print(U)
print("\nДиагональная матрица S:")
print(np.diag(S))
print("\nОртогональная матрица Vt:")
print(Vt)
result = U[:8, :8]@S@Vt
print("\nРезультат умножения U, S и Vt:")
print(result)


```

- Необходимо вычислить собственные векторы и собственные значения матрицы **X_matrix** при помощи numpy:
    - Создайте перменную **w** и присвойте ей массив с собственными значениями матрицы X_matrix (тип numpy.ndarray)
    - Создайте перменную **v** и присвойте ей матрицу с собственными векторами матрицы X_matrix (тип numpy.ndarray)
- Проверьте, что при умножении матрицы X_matrix и собственного вектора, получается тот же результат, что и при умножении собственного значения и собственного вектора, для этого:
    - Создайте переменную **check_1** и присвойте ей результат перемножения матрицы X_matrix на нулевой вектор из **v**
    - Создайте переменную **check_2** и присвойте ей результат перемножения нулевого вектора **v** на собственное значение **w** под индексом 0
    
**Подсказка:**

Получившийся результат в **check_1** должен быть равен **check_2**

```python
from numpy import linalg as LA
X_matrix = np.array([[-1, -6], [2, 6]])
w, v = LA.eig(X_matrix)

check_1 = X_matrix@v[:, 0]
check_2 = w[0]*v[:, 0]
print("Собственные значения w:")
print(w)
print("\nСобственные векторы v:")
print(v)
print("\nПеремножение X_matrix на собственный вектор:")
print(check_1)
print("\nПеремножение собственного значения на собственный вектор:")
print(check_2)
```

### HW. Блок Numpy. Задача 6

- Вам даны два вектора **a** и **b** ниже (скопируйте код и вставьте в code editor)
- Создайте переменную **similarity** и присвойте значение косинуса угла между векторами **a** и **b** (можно использовать scipy)
- найдите длины векторов:
    - Создайте переменную **dist_a_manh** и присвойте значение Манхэттенской нормы вектора **a**
    - Создайте переменную **dist_b_manh** и присвойте значение Манхэттенской нормы вектора **b**
    - Создайте переменную **dist_a_euclid** и присвойте значение Евклидовой нормы вектора **a**
    - Создайте переменную **dist_b_euclid** и присвойте значение Евклидовой нормы вектора **b**

```python
from scipy import spatial

a = np.array([2, 4, 5])
b = np.array([1, 3, 2])

# Ваш код

similarity = 1 - spatial.distance.cosine(a, b)
dist_a_manh = np.linalg.norm(a, ord=1)
dist_b_manh = np.linalg.norm(b, ord=1)
dist_a_euclid = np.linalg.norm(a, ord=2)
dist_b_euclid = np.linalg.norm(b, ord=2)
print("Косинус угла между векторами a и b:")
print(similarity)
print("\nМанхэттенская норма вектора a:")
print(dist_a_manh)
print("\nМанхэттенская норма вектора b")
print(dist_a_manh)
print("\nЕвклидова норма вектора a")
print(dist_a_euclid)
print("\nЕвклидова норма вектора b")
print(dist_b_euclid)
```
